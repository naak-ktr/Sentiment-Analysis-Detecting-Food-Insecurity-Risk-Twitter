**Plotting the WordCloud**

Creating the new list of tweet column from the DataFrame and giving it "sentences" name variable.

```python
# Creating the new list of sentences
sentences = tweets_df["Cleaned_Tweets"].tolist()
sentences[:5]
```

Creating a single string by joining all the elements of the "sentences" list.

**1. Word Cloud of Tweets with Positive Sentiments**

```python
# Create Positive Tweets Dataframe
positive = tweets_df[tweets_df["Harmonized_Label"] == 'Positive']
display(positive)
```

```python
from wordcloud import WordCloud

# Creating the WordCloud of the POSITIVE tweet DataFrame
sentences = positive["Cleaned_Tweets"].tolist()
sentences_string = " ".join(sentences)
plt.figure(figsize=(10, 10))
plt.imshow(WordCloud(background_color="white", max_words=1000).generate(sentences_string))
plt.axis('off')
```

```python
# WordCLoud in Twitter Shape
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

mask = np.array(Image.open("twitter_mask.png"))
cloud = WordCloud(background_color="white", max_words=200, random_state=42, mask=mask).generate(sentences_string)

plt.figure()
plt.imshow(cloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
# store to file
plt.savefig("Twitter Positive Sentiment Tweets (Harmonized).png", format="png")
plt.show()
```
